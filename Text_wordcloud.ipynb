```python
#https://www.kaggle.com/code/colearninglounge/nlp-data-preprocessing-and-cleaning/notebook

import nltk
from nltk.corpus import stopwords
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
nltk.download('stopwords')
```

```python
df = pd.read_csv("sample_data/IMDB Dataset.csv")
text_column = df['review']
```

# Nouvelle section

```python
text = " ".join(text_column)

mots_a_exclure = ["br"]

# Supprimez les mots à exclure du texte
for mot in mots_a_exclure:
    text = text.replace(mot, "")
```

```python
stop_words = set(stopwords.words('english'))  # Choisissez la langue correspondant à votre texte
words = text.split()
filtered_words = [word for word in words if word.lower() not in stop_words]
filtered_text = " ".join(filtered_words)
```

```python
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(filtered_text)
```

```python
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
